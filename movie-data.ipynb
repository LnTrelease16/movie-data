```python
# Import libraries
import urllib3   # allows access to urls via python
import certifi   # validates the trustworthiness of ssl certificates
import re        # allows python to read regualr expressions
# insert the url 
url = "https://www.metacritic.com/browse/movie/all/all/2020/metascore/?page=1"
# construct a pool manager to generate request
http = urllib3.PoolManager(ca_certs=certifi.where())
# initiates a web request
r = http.request('GET', url, headers={'user-agent': 'Mozilla/5.0'})
# converts bytes to text
datastring = str(r.data, "utf-8")
#check results
print(r.status)
print(len(datastring))
```

```python
# defining regular expression
title_regularexpression = r'<div data-title="(.*?)"'

# finds matches for the regular expression and stores them in an array 
title_matches = re.findall(title_regularexpression, datastring)

# prints out results
print(f"Found {len(title_matches)} matches. The first match is {title_matches[0]}")
```

```python
# imports pandas 
import pandas as pd
# imports style module from pandas to display the text nicely
from pandas.io.formats import style
# creates a pandas dataframe
data = {
    "Title": title_matches
    }
dataframe = pd.DataFrame(data)
# formats and displays results nicely
left_aligned_df = dataframe.style.set_properties(**{'text-align': 'left'})
display(left_aligned_df)
```

```python
# defining regular expression
release_date_regular_expression = r'class="u-text-uppercase">\s*(.*)\s'

# finds matches for the regular expression and stores them in an array 
release_date__matches = re.findall(release_date_regular_expression, datastring)

# prints out results
print(f"Found {len(release_date__matches)} matches. The first match is {release_date__matches[0]}")
```

```python
# creates a pandas dataframe
data = {
    "Release Date": release_date__matches
    }
dataframe = pd.DataFrame(data)

# formats and displays results nicely
left_aligned_df = dataframe.style.set_properties(**{'text-align': 'left'})
display(left_aligned_df)
```

```python
# defining regular expression
metascore_regular_expression = r'data-v-4cdca868><div title="\w*\s(.*?)\s'

# finds matches for the regular expression and stores them in a array
metascore_matches = re.findall(metascore_regular_expression, datastring)

# prints out results
print(f"Found {len(metascore_matches)} matches. The first match is {metascore_matches[0]}")
```

```python
# creates a pandas data frame
data = {
    "Metascore": metascore_matches
    }
dataframe = pd.DataFrame(data)

# formats and displays results nicely
left_aligned_df = dataframe.style.set_properties(**{'text-align': 'left'})
display(left_aligned_df)
```

```python
# defining regular expression
thumbnail_regular_expression = r'<img src="(.*?\d)"'

# finds matches for our regular expression and stores them in an array
thumbnail_matches = re.findall(thumbnail_regular_expression, datastring)

# fixing url
fixed = []
for url in thumbnail_matches:
    url = url.replace("&amp;", "&")
    fixed.append(url)

print(f"Found {len(fixed)} matches. First match is {fixed[0]}")
```

```python
# imports style module from pandas to display the text nicely
from pandas.io.formats import style

# creates a pandas data frame
data = {
    "Thumbnail": fixed
    }
dataframe = pd.DataFrame(data)
# display results
left_aligned_df = dataframe.style.set_properties(**{'text-align': 'left'})
display(left_aligned_df)
```

```python
# defining regular expression
description_regular_expression = r'<div class="c-finderProductCard_description"><span>([^>]*)<'

# finds matches and stores them in an array
description_matches = re.findall(description_regular_expression, datastring)

# prints out results
print(f"Found {len(description_matches)} matches. First match is {description_matches[0]}")
```

```python
# imports style module from pandas to display the text nicely
from pandas.io.formats import style

# creates a pandas dataframe
data = {
    "Description": description_matches
    }
dataframe = pd.DataFrame(data)

# aligns the text on the left side of the dataframe
left_aligned_df = dataframe.style.set_properties(**{'text-align': 'left'})
display(left_aligned_df)
```
